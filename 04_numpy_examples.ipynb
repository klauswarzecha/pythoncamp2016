```python
import numpy as np
```

```python
data='022-orig-abs.txt'
spectrum = np.loadtxt(data)
spectrum[:3]
```

*spectrum* is ein Array aus Wertepaaren (Wellenlänge, Absorbanz)

```python
wavelength, absorbance = spectrum = np.loadtxt(data, unpack=True)
print(wavelength[:3], absorbance[:3])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure()
plt.plot(wavelength, absorbance)
```

```python
lower = wavelength > 280
upper = wavelength < 400
fromto = lower & upper

wavelength_ok = wavelength[fromto]
absorbance_ok = absorbance[fromto]
```

```python
plt.figure(figsize=(8,8))
plt.plot(wavelength_ok, absorbance_ok, 'r-')
```

```python
nice_spec = np.array([wavelength_ok, absorbance_ok]).T
```

```python
np.savetxt('anthracene_abs', nice_spec, header='anthracene')
```
